```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df['Investors Name']=df['Investors Name'].fillna('undisclosed')
```

```python
df['Remarks'].info()
```

```python
df=df.drop(columns=['Remarks','Sr No'])
df.head()
```

```python
df=df.rename(columns={
    'Date dd/mm/yyyy':'date',
    'Industry Vertical':'vertical',
    'City  Location':'city',
    'InvestmentnType':'round',
    'Investors Name':'investors'
})
df.head()
```

```python
df['Amount in USD'].info()
```

```python
usd_to_inr_rate = 82
df['Amount in USD'] = df['Amount in USD'].fillna('0')
df['Amount in USD'] = df['Amount in USD'].str.replace('undisclosed', '0')
df['Amount in USD'] = df['Amount in USD'].str.replace(',', '')
df['Amount in USD'] = df['Amount in USD'].str.replace('unknown', '0')
df['Amount in USD'] = df['Amount in USD'].str.replace('Undisclosed', '0')
df = df[df['Amount in USD'].str.isdigit()]
df['Amount in USD'] = df['Amount in USD'].astype('float')

```

```python
def to_inr(dollar):
    inr=dollar*82
    return inr/10000000
```

```python
df['Amount in USD']=df['Amount in USD'].apply(to_inr)
df=df.rename(columns={'Amount in USD':'amount in cr(INR)'})
df.head()
```

```python
df.shape
```

```python
df['date'] = df['date'].str.replace('05/072018','05/07/2018')
df['date'] = df['date'].str.replace('01/07/015','01/07/2015')
df['date'] = df['date'].str.replace("12/05.2015","12/05/2015")
```

```python
df['date'].info()
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y',errors='coerce')
```

```python
df.shape
```

```python
df=df.dropna(subset=['date','Startup Name','vertical','city','round','amount in cr(INR)','investors'])
```

```python
df.to_csv('startup_cleaned',index=False)
```
